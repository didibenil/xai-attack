Load and explore the Kaggle Credit Card Fraud dataset.

```python
import sys
print(f"Python: {sys.executable}")
import pandas as pd
print(f"Pandas: {pd.__version__}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification

# Generate synthetic fraud dataset (harder to separate than real credit card data)
print("Generating synthetic fraud detection dataset...")

X, y = make_classification(
    n_samples=100000,           # 100k transactions
    n_features=30,              # 30 features
    n_informative=10,           # Only 10 features truly matter
    n_redundant=10,             # 10 redundant features
    n_repeated=5,               # 5 repeated features
    n_classes=2,
    weights=[0.995, 0.005],     # 0.5% fraud rate (imbalanced)
    random_state=42,
    flip_y=0.01                 # 1% label noise (real-world confusion)
)

# Create DataFrame
feature_names = [f'V{i}' for i in range(30)]
df = pd.DataFrame(X, columns=feature_names)
df['Class'] = y

print(f"Dataset shape: {df.shape}")
print(f"\nFirst few rows:")
print(df.head())
print(f"\nData info:")
print(df.info())
```

## Basic Statistics

```python
print(f"\nClass distribution:")
print(df['Class'].value_counts())
print(f"\nFraud percentage: {df['Class'].sum() / len(df) * 100:.2f}%")

print(f"\nFeature statistics:")
print(df.describe())
```

## Check for Missing Values and Correlations

```python
# Missing values
print(f"Missing values:\n{df.isnull().sum()}")

# Visualize correlation with target
plt.figure(figsize=(12, 6))
correlation_with_target = df.corr()['Class'].sort_values(ascending=False)
print(f"\nTop features correlated with fraud:")
print(correlation_with_target.head(10))
```
