# 5. Attack WITHOUT XAI (Baseline)

Execute adversarial attack using random feature manipulation (no explanations).

```python
import pandas as pd
import numpy as np
import pickle
import sys
sys.path.insert(0, '../src')
from attack_without_xai import RandomFeatureAttack

# Load model, data, and scaler
with open('../data/model.pkl', 'rb') as f:
    model = pickle.load(f)

with open('../data/scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)

with open('../data/feature_names.pkl', 'rb') as f:
    feature_names = pickle.load(f)

# Load original data
df = pd.read_csv('../data/creditcard.csv')
X = df.drop('Class', axis=1).values
X_scaled = scaler.transform(X)
y = df['Class'].values

print(f"Data shape: {X_scaled.shape}")
```

## Execute Random Feature Attack

```python
# Initialize attacker WITHOUT SHAP access (random baseline)
attacker = RandomFeatureAttack(
    model=model,
    feature_names=feature_names,
    target_class=0,
    seed=42
)

# Run attack on test set (same data as WITH XAI for fair comparison)
print("Executing random feature attack...")
results_without_xai = attacker.attack(
    X=X_scaled,
    y_true=y,
    max_iterations=50,
    step_size=0.05
)

print("Attack complete!")
```

## Attack Results

```python
print("\n=== ATTACK WITHOUT XAI (Random) ===")
print(f"Total targets: {results_without_xai['successful_attacks'] + results_without_xai['failed_attacks']}")
print(f"Successful attacks: {results_without_xai['successful_attacks']}")
print(f"Failed attacks: {results_without_xai['failed_attacks']}")
print(f"Success rate: {results_without_xai['success_rate']:.2%}")
print(f"\nAvg features modified: {results_without_xai['avg_features_modified']:.2f}")
print(f"Avg modification magnitude: {results_without_xai['avg_modification_magnitude']:.4f}")
print(f"Total model queries: {results_without_xai['queries']}")

# Save results
with open('../data/results_without_xai.pkl', 'wb') as f:
    pickle.dump(results_without_xai, f)

print("\nResults saved!")
```
