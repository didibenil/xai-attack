# 4. Attack WITH XAI (SHAP-Guided)

Execute adversarial attack using SHAP explanations to guide feature manipulation.

```python
import pandas as pd
import numpy as np
import pickle
import sys
sys.path.insert(0, '../src')
from attack_with_xai import SHAPGuidedAttack

# Load model, data, and scaler
with open('../data/model.pkl', 'rb') as f:
    model = pickle.load(f)

with open('../data/scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)

with open('../data/feature_names.pkl', 'rb') as f:
    feature_names = pickle.load(f)

# Load SHAP values
shap_values = np.load('../data/shap_values.npy')

# Load original data
df = pd.read_csv('../data/creditcard.csv')
X = df.drop('Class', axis=1).values
X_scaled = scaler.transform(X)
y = df['Class'].values

print(f"Data shape: {X_scaled.shape}")
print(f"SHAP values shape: {shap_values.shape}")
```

## Execute SHAP-Guided Attack

```python
# Initialize attacker with SHAP access
attacker = SHAPGuidedAttack(
    model=model,
    shap_values=shap_values,
    feature_names=feature_names,
    target_class=0  # Target: legitimate (class 0)
)

# Run attack on test set
print("Executing SHAP-guided attack...")
results_with_xai = attacker.attack(
    X=X_scaled,
    y_true=y,
    max_iterations=50,
    top_k=5,
    step_size=0.05
)

print("Attack complete!")
```

## Attack Results

```python
print("\n=== ATTACK WITH XAI (SHAP-Guided) ===")
print(f"Total targets: {results_with_xai['successful_attacks'] + results_with_xai['failed_attacks']}")
print(f"Successful attacks: {results_with_xai['successful_attacks']}")
print(f"Failed attacks: {results_with_xai['failed_attacks']}")
print(f"Success rate: {results_with_xai['success_rate']:.2%}")
print(f"\nAvg features modified: {results_with_xai['avg_features_modified']:.2f}")
print(f"Avg modification magnitude: {results_with_xai['avg_modification_magnitude']:.4f}")
print(f"Total model queries: {results_with_xai['queries']}")

# Save results
with open('../data/results_with_xai.pkl', 'wb') as f:
    pickle.dump(results_with_xai, f)

print("\nResults saved!")
```
